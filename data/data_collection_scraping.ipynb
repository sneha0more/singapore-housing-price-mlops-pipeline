```python
# Create a dictionary that stores features to be collected
property_details = {
    "title": [],
    "price": [],
    "n_rooms": [],
    "area": [],
    "year": [],
    "price_per_sqft": [],
    "size_sqft": [],
    "road_name": [],
    "n_bedrooms": [],
    "n_bathrooms": [],
    "region": []
    }  
```

```python
def scrape_properties(pages):

    for k,v in urls.items(): 
        region = k
        base_url = v

        for pg in range(1, pages+1): 

            url = base_url.format(page=pg)
            driver.get(url)

            # Getting title and price, description boxes, the numbers of bedrooms and bathrooms for each property 
            # Wait for at least one title to appear
            try:
                WebDriverWait(driver, 10).until(EC.presence_of_element_located(
                        (By.XPATH, "//div[contains(@class, 'listing-card-name')]/h2")
                    )
                )
            except:
                print(f"No listings found on page or took too long to load.")

            xpath_title = '//div[@class="jsx-911604640 listing-card-name"]/h2' 
            xpath_price = '//div[@class="jsx-911604640 listing-card-price"]/a'
            xpath_descs = '//div[contains(@class, "jsx-4147696354 desc-box")]'
            xpath_nrooms = './/span[@class="jsx-609329512"]/span'

            elements_title = driver.find_elements(By.XPATH, xpath_title)
            elements_price = driver.find_elements(By.XPATH, xpath_price)
            elements_desc = driver.find_elements(By.XPATH, xpath_descs)
            elements_nrooms = driver.find_elements(By.XPATH, xpath_nrooms)
            
            print(f"Page {pg}: Found {len(elements_title)} titles, {len(elements_price)} prices, {len(elements_desc)} desc-boxes.")
            
            # Checking the coherence between number of titles and prices collected
            if (len(property_details['title'])) != (len(property_details['price'])):
                print('Warning: Number of titles and prices are different')

            # Find the cards that contains the listing details for the 'n_bedrooms','n_bathrooms'. 
            try:
                WebDriverWait(driver, 10).until(EC.presence_of_element_located(
                        (By.XPATH, "//div[contains(@class, 'listing-card-detail')]")
                    )
                )
            except:
                print(f"took too long to load.")
            
            cards = driver.find_elements(By.XPATH, "//div[contains(@class, 'listing-card-detail')]")
                
            start_idx = 0

            desc_texts = [el.text for el in elements_desc]

            # Detect property start indices based on first attribute containing expected keyword
            # HDB 5-Room  
            start_indices = []
            for idx, text in enumerate(desc_texts):
                if "HDB" in text: 
                    start_indices.append(idx)
                    
            # Use these indices to chunk property descriptions
            for i in range(len(elements_title)):
                if i >= len(start_indices):
                    break

                start_idx = start_indices[i]
                end_idx = start_indices[i + 1] if i + 1 < len(start_indices) else len(desc_texts)
                chunk = desc_texts[start_idx:end_idx] # Store the attributes

                if len(chunk) < 5:  # at least 5 attributes to proceed
                    continue

                # Get n_room info 
                try:
                    elements_nrooms = cards[i].find_element(By.XPATH, xpath_nrooms)
                    text = elements_nrooms.get_attribute('textContent') # Example output: "3 beds | 2 baths"
                    text_split = [t.strip() for t in text.split("|")]  #["3 beds", "2 baths"]
                    if len(text_split) != 2:
                        continue
                except:
                    continue

                # Map attributes to fields
                property_details['title'].append(elements_title[i].text)
                property_details['price'].append(elements_price[i].text)
                property_details['n_rooms'].append(chunk[0])
                property_details['area'].append(chunk[1])
                property_details['year'].append(chunk[2])
                property_details['price_per_sqft'].append(chunk[3] if len(chunk) > 3 else "")
                property_details['size_sqft'].append(chunk[4] if len(chunk) > 4 else "")
                property_details['road_name'].append(chunk[5] if len(chunk) > 5 else "")
                property_details['n_bedrooms'].append(text_split[0])
                property_details['n_bathrooms'].append(text_split[1])
                property_details['region'].append(region)


    return property_details
```

```python
# Use DataFrame format to check 
##FOR TESTING OR CHECKING, PLEASE CHANGE THE PAGES ACCORDINGLY 

property_details = scrape_properties(pages=10)

df = pd.DataFrame.from_dict(property_details)

print('Number of listings retrieved:', len(df['title']))
```

```python
## Adding timestamp for time features
from datetime import datetime, timedelta
import random

# Set start and end date range
start_date = datetime(2025, 2, 25)
end_date = datetime(2025, 3, 27)
date_range = pd.date_range(start=start_date, end=end_date)

# Generate timestamps 
dates = []
for i in range(0, len(df), 10):
    # Randomly pick a date from the range for this batch
    batch_date = random.choice(date_range)
    batch_size = min(10, len(df) - i)
    dates.extend([batch_date] * batch_size)

# Assign to the dataframe
df['scraped_date'] = dates
```
