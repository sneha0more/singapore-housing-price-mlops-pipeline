###Feature Engineering

```python
df['scraped_date'] = pd.to_datetime(df['scraped_date'], errors='coerce')
df['scraped_date'].describe()
```

```python
df = df.rename(columns={'year': 'build_year'})
```

```python
# --- Feature Engineering ---

# 1. Flat Age
df['flat_age'] = 2025 - df['build_year']

# 2. Lease Remaining (assuming 99-year HDB lease)
df['lease_remaining'] = 99 - df['flat_age']

# 3. Price per Bedroom
df['price_per_bedroom'] = df['price'] / df['n_bedrooms']

# 5. Size per Room
df['size_per_room'] = df['size_sqft'] / df['n_bedrooms']

# 6. Bedroom to Bathroom Ratio
df['bed_bath_ratio'] = df['n_bedrooms'] / df['n_bathrooms']

# 7. Extract scraped month and year
df['scraped_date'] = pd.to_datetime(df['scraped_date'])
df['scraped_month'] = df['scraped_date'].dt.month
df['scraped_year'] = df['scraped_date'].dt.year

# 8. (Optional) Drop unused columns if needed
df = df.drop(columns=["Unnamed: 0", "Unnamed: 0.1"])

#For better visibility to the model, need to flag it
df['is_central'] = (df['region'].str.lower() == 'central').astype(int)

# Define mature towns in Singapore
mature_towns = [
    'Ang Mo Kio', 'Bedok', 'Bishan', 'Bukit Merah', 'Bukit Timah', 'Central Area',
    'Clementi', 'Geylang', 'Kallang/Whampoa', 'Marine Parade',
    'Pasir Ris', 'Queenstown', 'Serangoon', 'Tampines', 'Toa Payoh'
]

# Add a binary feature: 1 if town is mature, 0 otherwise
df['is_mature_town'] = df['area'].isin(mature_towns).astype(int)

#This captures how older + larger flats behave in pricing.
	#	Some old flats may still be big and valuable.
	#	Others may be old and small → lower value.
df['age_size_interaction'] = df['flat_age'] * df['size_sqft']

# This tells the model:
	# A flat with 90 years lease and $800 psf is very different from one with 30 years at $800 psf.
	# Lease affects how much the per sqft price matters.
df['lease_price_interaction'] = df['lease_remaining'] * df['price_per_sqft']


# Display the first few rows
print(df.head())
```
